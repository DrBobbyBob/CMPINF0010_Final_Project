The best neighborhood is a neighborhood with the highest median age of death. This statstic is usually an indicator of the healthcare system, violent crime rates, and quality of life in the neighborhood. These traits are highly sought after in prospective homeowners. Therefore the higher the median age of death mean as better neighborhood to live in.

This code segement is removing non-valid neighborhoods with either no MD death age or a MD death age of 0: South Shore, Chateau, and a couple empty categories.

```python
ageDeathData =  pd.read_csv("nhoodmedianage20112015.csv")
querymask = ageDeathData['TOTAL MD AGE AT DEATH'] > 0
```

This is all of the neighborhoods ranked from the lowest median age of death to the highest median age of death.

```python
mdDeath = ageDeathData[querymask]

df = mdDeath[["NEIGHBORHOOD","TOTAL MD AGE AT DEATH"]]

sorted_df = df.sort_values(by='TOTAL MD AGE AT DEATH')

sorted_df.plot(x='NEIGHBORHOOD', y='TOTAL MD AGE AT DEATH', kind="bar", figsize=(20,8))
```

Here is the 10 neighborhoods with the highest median age of death:

```python
highest = sorted_df.tail(10)
highest.plot(x='NEIGHBORHOOD', y='TOTAL MD AGE AT DEATH', kind="bar", figsize=(20,8))
```
